```python
from sqlalchemy import create_engine, Column, Integer, String, Numeric, Date, DateTime, ForeignKey
from sqlalchemy.orm import declarative_base, relationship
from sqlalchemy.sql import func
from sqlalchemy_utils import database_exists, create_database

# -- Variables globales -- #
DBMS = 'postgresql'
DRIVER = 'psycopg2'
USER = 'default'
PASSWORD = 'K1PgwR4caFev'
HOST = 'ep-summer-frog-90198346-pooler.us-east-1.postgres.vercel-storage.com'
PORT = '5432'
DB_NAME = 'verceldb'


# -- Creaci贸n del engine y de la db -- #
engine = create_engine(f'{DBMS}://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB_NAME}')
if not database_exists(engine.url):
    create_database(engine.url)

Base = declarative_base()  # Usada para declarar y crear las tablas

# -- Declaraci贸n de tablas -- #
    
class Product(Base):
    __tablename__ = 'product'
    product_id = Column(Integer, primary_key=True, autoincrement=True)
    product_name = Column(String(150))
    product_cat = Column(String(150))
    product_subcat = Column(String(150))
    product_container = Column(String(150))
    product_unit_price = Column(Numeric(9,2))
    product_ship_cost = Column(Numeric(9,2))
    product_base_margin = Column(Numeric(9,2))

    order = relationship("order")
        

class Customer(Base):
    __tablename__ = 'customer'
    customer_id = Column(Integer, primary_key=True)
    customer_name = Column(String(150))

    order = relationship("order")
    


class Location(Base):
    __tablename__ = 'location'
    country = Column(String(150))
    region = Column(String(150))
    state = Column(String(150))
    city = Column(String(150))
    postal_code = Column(Integer, primary_key=True)

    order = relationship("order")


class Order(Base):
    __tablename__ = 'order'
    order_id = Column(Integer, primary_key=True)
    customer_id = Column(Integer, ForeignKey("customer.customer_id"))
    product_id = Column(Integer, ForeignKey("product.product_id"))
    postal_code = Column(Integer, ForeignKey("location.postal_code"))
    order_prio = Column(String(150))
    ship_mode = Column(String(150))
    discount = Column(Numeric(9,2))
    order_date = Column(Date)
    ship_date = Column(Date)
    customer_segment = Column(String(150))
    sales = Column(Numeric(9,2))
    profit = Column(Numeric(9,2))
    qty_ordered = Column(Integer)
    row_id = Column(Integer, primary_key=True)
    time_created = Column(DateTime(timezone=True), server_default=func.now())


class Temp(Base):
    __tablename__ = 'temp'
    row_id = Column(Integer)
    order_prio = Column(String(150))
    discount = Column(Numeric(9,2))
    unit_price = Column(Numeric(9,2))
    ship_cost = Column(Numeric(9,2))
    customer_id = Column(Integer)
    customer_name = Column(String(150))
    ship_mode = Column(String(150))
    customer_segment = Column(String(150))
    product_cat = Column(String(150))
    product_subcat = Column(String(150))
    product_cont = Column(String(150))
    product_name = Column(String(150))
    product_base_margin = Column(Numeric(9,2))
    country = Column(String(150))
    region = Column(String(150))
    state = Column(String(150))
    city = Column(String(150))
    postal_code = Column(Integer)
    order_date = Column(String(150))
    ship_date = Column(String(150))
    profit = Column(Numeric(9,2))
    qty_ordered = Column(String(150))
    sales = Column(Numeric(9,2))
    order_id = Column(Integer)
    id = Column(Integer, primary_key=True, autoincrement=True)


# -- Creaci贸n de tablas en la DB -- #
Base.metadata.create_all(engine)

# Cerramos la conexi贸n
engine.dispose()
```
